First function: add some value to a chosen column
Second function: add some value to a chosen row
Third function: find the smallest value in the grid

```python
def addColumn(col, num, grid):
    for row in grid:
        row[col-1]+num
    return grid

def addRow(row, num, grid):
    for i in range(len(grid(row-1]))):
        grid[row-1][i] +=num
    return grid

def minimum(grid):
    mins= []
    for row in grid:
        mins.append(min(row))
    return grid


grid = []
for i in range(5):
    grid.append([0, 0, 0, 0, 0])
grid = addRow(1,1,grid)

for g in grid:
    print(g)


    
```
