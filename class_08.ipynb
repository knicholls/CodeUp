```python
def print_grid(g):
    for i in g:
        print(i)

g = [[" " for _ in range(5)] for _ in range(5)]
n = int(input("Hw many changes would you like to make to the table? "))
for i in range(n):
        row = int(input("Input the row number: "))
        col = int(input("Input the column number: "))
        s = input("Enter the cells new value: ")
        g[row - 1][col - 1] = s

        print_grid(g)
```

Second function: add some value to a chosen row
Third function: find the smallest value in the grid

```python
def addColumn(col, num, grid):
    for row in grid:
        row[col-1]+num
    return grid

def addRow(row, num, grid):
    for i in range(len(grid(row-1]))):
        grid[row-1][i] +=num
    return grid

def minimum(grid):
    mins= []
    for row in grid:
        mins.append(min(row))
    return grid


grid = []
for i in range(5):
    grid.append([0, 0, 0, 0, 0])
grid = addRow(1,1,grid)

for g in grid:
    print(g)


    
```
